```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import skimage.exposure
import imread
import h5py
import glob
import time
import scipy.ndimage as nd

from matplotlib import rcParams
rcParams['figure.figsize'] = (10, 10)
rcParams['figure.dpi'] = 300
```

```python
window_size = 19
features = {}    
```

```python
def show_subimage(im):
    plt.figure()
    plt.imshow(im[:1024, :1024], cmap=plt.cm.gray)

# use a subset for testing purposes
image_in = imread.imread('04.tif')

print image_in.shape
show_subimage(image_in)
features['original'] = image_in
```

**Preprocessing: Adaptive histogram equalization**

```python
adapt_im = skimage.exposure.equalize_adapthist(image_in, clip_limit=0.01, nbins=256) / 256
print adapt_im.max(), adapt_im.dtype
adapt_im = adapt_im.astype(np.uint8)
show_subimage(adapt_im)
features['adapthisteq'] = image_in
```

**Feature: weighted sum of pixels in each decile**

```python
s = 0
for decile in range(10):
    lo_thresh = (decile * 256) / 10.0
    hi_thresh = ((decile + 1) * 256) / 10.0
    thresholded_im = np.logical_and(adapt_im >= lo_thresh, adapt_im < hi_thresh).astype(np.uint8) * 255
    features['decile_%d' % decile] = nd.filters.gaussian_filter(thresholded_im, window_size / 2.0).astype(np.uint8)
    
    
show_subimage(features['decile_9'])
show_subimage(features['decile_2'])
show_subimage(features['decile_5'])
show_subimage(features['decile_8'])
```

Feature: **Difference of Gaussian (2 minus 50)**

```python
DoG = (nd.filters.gaussian_filter(adapt_im, 2).astype(float) -
       nd.filters.gaussian_filter(adapt_im, 50).astype(float))
DoG += 255
DoG /= 2.0
features['DoG_2_50'] = DoG.astype(np.uint8)
show_subimage(DoG)
```

**Features: Smoothed local mean and variance**

```python
local_mean = nd.filters.gaussian_filter(adapt_im, window_size)
local_variance = nd.filters.gaussian_filter(adapt_im ** 2.0, window_size) - local_mean ** 2
local_variance = (local_variance / 255.0).astype(np.uint8)
features['local_mean'] = local_mean
features['local_variance'] = local_variance
show_subimage(local_mean)
show_subimage(local_variance)
```

**Feature: Blurred image with sigma = 10**

```python
blur_10 = nd.filters.gaussian_filter(adapt_im, 10)
show_subimage(blur_10)
features['blur_10'] = blur_10
```

```python
def compute_features(image_in, window_size=19):
    features = {}

    # Feature: original image
    features['original'] = image_in

    # Feature: adaptive histogram equalization
    adapt_im = skimage.exposure.equalize_adapthist(image_in, clip_limit=0.01, nbins=256) / 256
    adapt_im = adapt_im.astype(np.uint8)
    features['adapthisteq'] = adapt_im
    
    # Feature: fraction of pixels in a given decile
    for decile in range(10):
        lo_thresh = (decile * 256) / 10.0
        hi_thresh = ((decile + 1) * 256) / 10.0
        thresholded_im = np.logical_and(adapt_im >= lo_thresh, adapt_im < hi_thresh).astype(np.uint8) * 255
        features['decile_%d' % decile] = nd.filters.gaussian_filter(thresholded_im, window_size / 2.0).astype(np.uint8)
    
    # Feature: Difference of Gaussians for sigma = 2, 50
    DoG = (nd.filters.gaussian_filter(adapt_im, 2).astype(float) -
           nd.filters.gaussian_filter(adapt_im, 50).astype(float))
    DoG += 255
    DoG /= 2.0
    features['DoG_2_50'] = DoG.astype(np.uint8)
    
    # Features: local mean and variance
    local_mean = nd.filters.gaussian_filter(adapt_im, window_size)
    local_variance = nd.filters.gaussian_filter(adapt_im ** 2.0, window_size) - local_mean ** 2
    local_variance = (local_variance / 255.0).astype(np.uint8)
    features['local_mean'] = local_mean
    features['local_variance'] = local_variance

    # Feature: image blurred with sigma 10
    features['blur_10'] = nd.filters.gaussian_filter(adapt_im, 10)
    
    return features
```

```python
input_files = glob.glob('??.tif')
for fname in input_files:
    start = time.time()
    features = compute_features(imread.imread(fname))
    print time.time() - start, "seconds to compute features for", fname
    output = h5py.File(fname[:2] + '.hdf5', 'w')
    for feature_name, data in sorted(features.items()):
        ds = output.create_dataset(feature_name, data=data, chunks=(128, 128), compression='szip')
    output.close()
```
