```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import imread
import h5py
import glob
import time
import sklearn.ensemble
import cPickle
import slimgraph
import structure
import scipy
from skimage.morphology import binary_opening, watershed, disk
from scipy import ndimage

from matplotlib import rcParams
rcParams['figure.figsize'] = (20, 20)
rcParams['figure.dpi'] = 300
```

```python
# Some values for the multisegmentations
opening_radii = [5, 11]
probexps = [.4, .2]
smoothness_weight = 0.1
gap_completion_weights = [0.02, 0.05, 0.1]
```

```python
with open('random_forest.pickle', 'rb') as fid:
    forest = cPickle.load(fid)
```

```python
def solve_graph(im, probabilities, similar_pixel_sigma, smoothness_weight, gap_completion_weight, opening_radius):
    num_nodes = im.size
    node_indices = np.arange(num_nodes, dtype=np.uint32).reshape(im.shape)
    graph = slimgraph.SlimGraph(num_nodes)
    # source = membrane, sink = background
    # Add foreground/background probabilities
    graph.add_source_capacities(np.minimum(0, 2 * (probabilities.ravel() - 0.5)))
    graph.add_sink_capacities(np.minimum(0, - 2 * (probabilities.ravel() - 0.5)))

    # Add a smoothness term that keeps similar intensities linked
    smooth = ndimage.gaussian_filter(im.astype(np.float32), 1)
    left_indices = node_indices[:, 1:]
    right_indices = node_indices[:, :-1]
    deltas = np.diff(smooth, axis=1)
    weights = np.exp(-((deltas ** 2) / (2 * sigma ** 2)))
    graph.add_edge_capacities(left_indices.ravel(), right_indices.ravel(), smoothness_weight * weights.ravel(), symmetric=True)
    
    up_indices = node_indices[:-1, :]
    down_indices = node_indices[1:, :]
    deltas = np.diff(smooth, axis=0)
    np.exp(-((deltas ** 2) / (2 * sigma ** 2)))
    graph.add_edge_capacities(up_indices.ravel(), down_indices.ravel(), smoothness_weight * weights.ravel(), symmetric=True)
    
    # Add a gap completion term that encourages linking in the direction of dominant texture
    concentration, eigx, eigy = structure.orientation_field(im, 2)
    concentration = concentration ** 2
    # eigx = 1.0 indicates that texture run vertically
    horizontal_weight = np.abs(gap_completion_weight * concentration * eigy)[:-1, :]
    vertical_weight = np.abs(gap_completion_weight * concentration * eigx)[:, :-1]
    dia_pos_weight = np.abs(gap_completion_weight * concentration * np.sqrt(2) * (eigx - eigy))[:-1, :-1]
    dia_neg_weight = np.abs(gap_completion_weight * concentration * np.sqrt(2) * (eigx + eigy))[:-1, :-1]
    graph.add_edge_capacities(left_indices.ravel(), right_indices.ravel(), horizontal_weight.ravel(), symmetric=True)
    graph.add_edge_capacities(up_indices.ravel(), down_indices.ravel(), vertical_weight.ravel(), symmetric=True)
    graph.add_edge_capacities(node_indices[:-1, :-1].ravel(), node_indices[1:, 1:].ravel(), dia_pos_weight.ravel(), symmetric=True)
    graph.add_edge_capacities(node_indices[1:, :-1].ravel(), node_indices[:-1, 1:].ravel(), dia_neg_weight.ravel(), symmetric=True)
    
    labels = graph.solve().reshape(im.shape)
    # clean up result
    labels = binary_opening(labels.astype(np.bool), disk(opening_radius))
    
    # Move region booundaries to watershed boundaries
    seeds = scipy.ndimage.label(labels)[0]
    blur_prob = (scipy.ndimage.gaussian_filter(probabilities.reshape(im.shape), 3.0) * (2**16 - 1)).astype(np.uint16)
    ws = watershed(blur_prob, seeds)
    dx, dy = np.gradient(ws)
    ws_boundary = np.logical_or(dx!=0, dy!=0)
    return ws_boundary
```

```python
# Show segmentations for different settings
opening_radius = 3
probexp = 0.4
gap_completion_weight = 0.02
result1 = solve_graph(im, probabilities ** probexp, sigma, smoothness_weight, gap_completion_weight, opening_radius)

opening_radius = 7
probexp = 0.4
gap_completion_weight = 0.02
result2 = solve_graph(im, probabilities ** probexp, sigma, smoothness_weight, gap_completion_weight, opening_radius)

plt.figure(dpi=300)
plt.subplot(1, 2, 1)
plt.imshow(result1, cmap=plt.cm.gray)
plt.title('open {} exp {} gap {}'.format(opening_radius, probexp, gap_completion_weight))
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(result2, cmap=plt.cm.gray)
plt.title('open {} exp {} gap {}'.format(opening_radius, probexp, gap_completion_weight))
plt.axis('off')
```

```python
# Generate multisegmentations for each image
input_files = glob.glob('??.tif')
for fname in input_files:
    start = time.time()
    im = imread.imread(fname)
    features = h5py.File(fname[:2] + '.hdf5', 'r')
    data_matrix = np.dstack([features[f][...].ravel() for f in sorted(features.keys())]).reshape(-1, 16)
    probabilities = forest.predict_proba(data_matrix)[:, 1].reshape(im.shape)

    # estimate variance of similar pixels using neighbor statistics
    local_diffs = np.diff(im.astype(np.float32), axis=0)
    MAD = np.median(abs(local_diffs - np.median(local_diffs)))
    sigma = 1.48 * MAD 
    print "SIGMA", sigma
    
    out_segmentations = h5py.File(fname[:2] + '_segs.hdf5', 'w')
    for opening_radius in [3, 7]:
        for probexp in [.4, .2]:
            for gap_completion_weight in [0.02, 0.05, 0.1]:
                result = solve_graph(im, probabilities ** probexp, sigma, 0.1, gap_completion_weight, opening_radius)
                segmentation_name = 'open_{}_gap_{}_exp_{}'.format(opening_radius, gap_completion_weight, probexp)
                out_segmentations.create_dataset(segmentation_name, data=result, chunks=(128, 128), compression='szip')
    print "Segmenting {} took {} seconds".format(fname, time.time() - start)
```
