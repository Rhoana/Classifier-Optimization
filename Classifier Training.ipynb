```python
def find_labeled_pixels(painted_image):
    R, G, B = [painted_image[:, :, idx] for idx in range(3)]
    red_pixels = R > np.maximum(G, B)
    green_pixels = G > np.maximum(R, B)
    blue_pixels = B > np.maximum(G, R)
    return red_pixels, green_pixels, blue_pixels

R, G, B = find_labeled_pixels(painted)

plt.subplot(1, 2, 1)
plt.imshow(painted)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(R, cmap=plt.cm.gray)
plt.axis('off')
```

```python
# Grab data from images

def extract_data_and_labels(painted_file, feature_file):
    R, G, B = find_labeled_pixels(imread.imread(painted_file))
    features = h5py.File(feature_file, 'r')
    num_features = len(features.keys())
    num_pixels = [np.sum(mask) for mask in [R, G, B]]
    offset = np.cumsum(num_pixels)
    
    data_matrix = np.empty((sum(num_pixels), num_features), dtype=np.uint8)
    for col, feature in enumerate(sorted(features.keys())):
        feature_data = features[feature][...]
        data_matrix[:, col] = np.hstack([feature_data[mask] for mask in [R, G, B]])
    label_matrix = np.zeros(sum(num_pixels), dtype=np.uint8)
    # Labels: RGB = 1, 2, 3
    label_matrix[:] = 2
    label_matrix[:num_pixels[0]] = 1
    label_matrix[-num_pixels[2]:] = 3
    print feature_file
    return data_matrix, label_matrix

painted_files = sorted(glob.glob('??_labeled.tif'))
feature_files = sorted(glob.glob('??.hdf5'))

data_matrices, label_matrices = zip(*(extract_data_and_labels(pf, ff) for pf, ff in zip(painted_files, feature_files)))
```

```python
def train_forest(n_trees, max_examples, data, labels):
    assert data.shape[1] == 16
    forest = sklearn.ensemble.RandomForestClassifier(n_estimators=n_trees, max_features=data.shape[1])
    subsample_mask = np.random.uniform(size=labels.shape) < (float(max_examples) / labels.shape[0])
    subtrain = data[subsample_mask.ravel(), :]
    sublabel = labels[subsample_mask]
    forest.fit(subtrain, sublabel)    
    return forest

def cross_validate(data_matrices, label_matrices, n_trees=20, max_examples=10000):
    errors = []
    for hold_out_idx in range(len(data_matrices)):
        training_data = np.vstack([data_matrices[idx] for idx in range(len(data_matrices)) if idx != hold_out_idx])
        training_labels = np.hstack([label_matrices[idx] for idx in range(len(data_matrices)) if idx != hold_out_idx]).reshape((-1, 1))
        forest = train_forest(n_trees, max_examples, training_data, training_labels)
        result = forest.predict(data_matrices[hold_out_idx]).ravel()
        truth = label_matrices[hold_out_idx].ravel()
        errors.append(np.sum(np.round(result) != truth) / float(truth.size))
    print "Errors (min/avg/max):", min(errors), np.mean(errors), max(errors)
    
cross_validate(data_matrices, label_matrices)
    
training_data = np.vstack([data_matrices[idx] for idx in range(len(data_matrices))])
training_labels = np.hstack([label_matrices[idx] for idx in range(len(data_matrices))]).reshape((-1, 1))
forest = train_forest(250, 100000, training_data, training_labels == 2)
print forest
```

```python
# save the classifier
with open('random_forest.pickle', 'wb') as fid:
    cPickle.dump(forest, fid)
```
